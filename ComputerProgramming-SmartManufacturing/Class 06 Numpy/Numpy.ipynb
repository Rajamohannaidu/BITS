```python
## Potential Energy Spring
```

```python

# Potential Energy (U) = 1/2 * k * X^2

# k --> Spring Constant (N/m)
# x --> Displacement (m)
```

```python
## Simulate Energy for Different Displacement
```

```python
x = np.linspace(0,10,50)

print(x)
```

```python
k = 100 ## spring constant
x = np.linspace(0,10,50)
U = 0.5 * k * (x**2)
print(U)
```

```python
import matplotlib.pyplot as plt

plt.plot(x,U, label = 'Potential Energy')
plt.xlabel("Displacement (m)")
plt.ylabel("Energy (Joules)")
plt.title("Sprint Potential Energy")
plt.grid(True)
plt.legend()
plt.show()
```

```python
if np.max(U)>3000:
  print("Warning: Spring Overload risk!")
```

```python
### Solving Linear Equations: Ax = b
```

```python
'''
2x + 3y = 8
3x + 4y = 11
'''
A = np.array([[2,3], [3,4]])
b = np.array([8,11])

x = np.linalg.solve(A,b)
print(x)
print(f"Solution: x = {x[0]}, y = {x[1]}")
```

```python
### Vectorization vs Loops
```

```python
## Looping
arr = [1,2,3,4,5]
squared = []
for i in arr:
  squared.append(i**2)
print(squared)
```

```python
## vectorization
arr_np = np.array([1,2,3,4,5])
squared_np = arr_np**2

print(squared_np)
```
