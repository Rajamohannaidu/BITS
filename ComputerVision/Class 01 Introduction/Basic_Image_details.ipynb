```python
import cv2
import matplotlib.pyplot as plt

# Load an image from a file
img = cv2.imread('bee-8790316_640.jpg')   # Replace with your own image path

# OpenCV loads images in BGR format; convert to RGB
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Show the image
plt.imshow(img_rgb)
plt.title("Original Image")
plt.axis('off')
plt.show()
```

```python
### The image is stored as a 3D NumPy array —
### each pixel is a combination of 3 values: Red, Green, and Blue.
```

```python
print("Shape of the image:", img_rgb.shape)   # (height, width, channels)
print("Pixel values at (0,0):", img_rgb[0,0])  # RGB values of top-left pixel
```

```python
## Here, we’re printing the actual matrix the computer sees. Each entry is a [R, G, B] triplet.

# Show a zoomed-in patch
patch = img_rgb[0:10, 0:10]   # Top-left 10x10 pixels
print("10x10 pixel values:\n", patch)
```

```python
### Grayscale images have only 1 value per pixel — intensity from black (0) to white (255).
### This is simpler and often used in early CV tasks.

img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

plt.imshow(img_gray, cmap='gray')
plt.title("Grayscale Image")
plt.axis('off')
plt.show()

print("Grayscale shape:", img_gray.shape)
print("Grayscale pixel value at (0,0):", img_gray[0,0])
```

```python
### This is how color is represented digitally — three separate layers stacked together (RGB).
### Each layer stores how much of that color to show.

r, g, b = img_rgb[:,:,0], img_rgb[:,:,1], img_rgb[:,:,2]

fig, axs = plt.subplots(1, 3, figsize=(12, 4))
axs[0].imshow(r, cmap='Reds')
axs[0].set_title('Red Channel')
axs[1].imshow(g, cmap='Greens')
axs[1].set_title('Green Channel')
axs[2].imshow(b, cmap='Blues')
axs[2].set_title('Blue Channel')

for ax in axs:
    ax.axis('off')
plt.show()
```
