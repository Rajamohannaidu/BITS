```python
import cv2
import matplotlib.pyplot as plt

# Load and convert to grayscale
img = cv2.imread('bee-8790316_640.jpg', cv2.IMREAD_GRAYSCALE)

# Generate Gaussian blurred images
gauss1 = cv2.GaussianBlur(img, (0, 0), sigmaX=1)
gauss2 = cv2.GaussianBlur(img, (0, 0), sigmaX=2)

# Compute DoG
DoG = gauss2 - gauss1

# Show results
plt.subplot(1, 3, 1), plt.imshow(gauss1, cmap='gray'), plt.title('Gaussian =1')
plt.subplot(1, 3, 2), plt.imshow(gauss2, cmap='gray'), plt.title('Gaussian =2')
plt.subplot(1, 3, 3), plt.imshow(DoG, cmap='gray'), plt.title('Difference of Gaussians')
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

# Read input image
img = cv2.imread('bee-8790316_640.jpg', cv2.IMREAD_GRAYSCALE)

# Create SIFT detector
sift = cv2.SIFT_create()

# Detect keypoints and compute descriptors
keypoints, descriptors = sift.detectAndCompute(img, None)

# Draw keypoints
img_keypoints = cv2.drawKeypoints(img, keypoints, None,
                                  flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)

# Show image
plt.imshow(img_keypoints, cmap='gray')
plt.title('SIFT Keypoints')
plt.axis('off')
plt.show()
```

```python
import numpy as np

H1 = np.array([2, 4, 3, 1])
H2 = np.array([1, 3, 5, 2])

# 1. L2 Distance
l2_dist = np.linalg.norm(H1 - H2)

# 2. Normalized Correlation
H1_mean = H1 - np.mean(H1)
H2_mean = H2 - np.mean(H2)
norm_corr = np.dot(H1_mean, H2_mean) / (np.linalg.norm(H1_mean) * np.linalg.norm(H2_mean))

# 3. Intersection
intersection = np.sum(np.minimum(H1, H2))

print("L2 Distance:", l2_dist)
print("Normalized Correlation:", norm_corr)
print("Intersection:", intersection)
```
