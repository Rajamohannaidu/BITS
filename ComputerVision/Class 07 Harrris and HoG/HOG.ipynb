```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
from skimage.feature import hog
from skimage import exposure

# 1. Create a tiny 5x5 grayscale image (intensity values)
image = np.array([
    [10, 10, 10, 10, 10],
    [10, 50, 50, 50, 10],
    [10, 50, 90, 50, 10],
    [10, 50, 50, 50, 10],
    [10, 10, 10, 10, 10]
], dtype='uint8')

plt.imshow(image, cmap='gray')
plt.title("Input 5x5 Grayscale Image")
plt.axis('off')
plt.show()

# 2. Compute HOG features (simple settings)
features, hog_image = hog(
    image,
    orientations=8,       # number of angle bins (e.g., 0â€“180 degrees)
    pixels_per_cell=(2, 2),  # small cell size
    cells_per_block=(1, 1),  # no normalization block
    visualize=True,
    feature_vector=True
)

# 3. Show HOG Image
plt.imshow(hog_image, cmap='gray')
plt.title("HOG Visualization")
plt.axis('off')
plt.show()

# 4. Print HOG Features
print("HOG Feature Vector:\n", np.round(features, 2))
```
